### CNN TIME

```python
import Preprocessor.PreprocessorTime as Preprocessor
X_train, y_train, X_val, y_val, X_test, y_test = Preprocessor.preprocess(True, True, True)
```

```python
print(X_train.shape)
```

Load CNN model

```python
from IA.CNN.TimeCNN import TimeCNN
time_cnn = TimeCNN(input_shape=(X_train[0].shape))
```

```python
time_cnn.compile()
```

```python
NR_EPOCHS = 100
BATCH_SIZE = 250
```

```python
time_cnn.fit(X_train, y_train, num_epochs=NR_EPOCHS, batch_size=BATCH_SIZE, val_data=X_val, val_labels=y_val)
```

```python
time_cnn.plot_train_val()
```

```python
time_cnn.predict(X_test)
time_cnn.print_predictions(y_test)
```
