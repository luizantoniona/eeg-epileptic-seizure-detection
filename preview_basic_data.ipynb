### Transformação domínio do tempo para frequência

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

N = 5000
T = 1.0 / 100.0
time = np.linspace(0.0, N*T, N)
frequency = np.linspace(0.0, 1.0/(2.0*T), N//2)

##Frequencies: 0.3Hz && 1Hz && 4Hz
temporal_wave = np.sin(0.3*2*np.pi*time) + 2*np.sin(2*np.pi*time) + np.sin(8*np.pi*time)
frequency_wave = np.fft.fft(temporal_wave)

figure = plt.figure(figsize=(5,5))

temporalx = figure.add_subplot(211)
temporalx.set_title('Temporal Waves')
temporalx.set_xlim(left=0, right=2)
temporalx.xaxis.set_major_locator(ticker.NullLocator())
temporalx.yaxis.set_major_locator(ticker.NullLocator())
temporalx.plot(time, temporal_wave)

frequencyx = figure.add_subplot(212)
frequencyx.set_title('Frequency Bands')
frequencyx.set_xlim(left=0, right=6)
frequencyx.xaxis.set_major_locator(ticker.NullLocator())
frequencyx.yaxis.set_major_locator(ticker.NullLocator())
frequencyx.plot(frequency, 2.0/N*np.abs(frequency_wave[:N//2]) )
```

### TFT e PSD

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.mlab as mlab

N = 50000
T = 1.0 / 500.0
time = np.linspace(0.0, N*T, N)
frequency = np.linspace(0.0, 1.0/(2.0*T), N//2)

delta_waves = np.sin(0.3*2*np.pi*time) + np.sin(1*2*np.pi*time) + np.sin(4*2*np.pi*time)
theta_waves = np.sin(4.5*2*np.pi*time) + np.sin(6*2*np.pi*time) + np.sin(8*2*np.pi*time)
alpha_waves = np.sin(8.5*2*np.pi*time) + np.sin(10*2*np.pi*time) + np.sin(13*2*np.pi*time)
beta_waves = np.sin(14*2*np.pi*time) + np.sin(20*2*np.pi*time) + np.sin(25*2*np.pi*time) + np.sin(30*2*np.pi*time)
lowgamma_waves = np.sin(31*2*np.pi*time) + np.sin(40*2*np.pi*time) + np.sin(45*2*np.pi*time) + np.sin(50*2*np.pi*time) + np.sin(80*2*np.pi*time)
highgamma_waves = np.sin(80.5*2*np.pi*time) + np.sin(90*2*np.pi*time) + np.sin(100*2*np.pi*time) + np.sin(125*2*np.pi*time) + np.sin(150*2*np.pi*time)

temporal_wave = delta_waves + theta_waves - alpha_waves + beta_waves + lowgamma_waves + highgamma_waves
frequency_wave = np.fft.fft(temporal_wave)

figure = plt.figure(figsize=(15,10))

temporalx = figure.add_subplot(311)
temporalx.set_title('Temporal Waves')
temporalx.set_xlim(left=0, right=2)
temporalx.xaxis.set_major_locator(ticker.NullLocator())
temporalx.yaxis.set_major_locator(ticker.NullLocator())
temporalx.plot(time, temporal_wave)

frequencyx = figure.add_subplot(312)
frequencyx.set_title('Frequency Bands')
frequencyx.set_xlim(left=0, right=150)
frequencyx.xaxis.set_major_locator(ticker.NullLocator())
frequencyx.yaxis.set_major_locator(ticker.NullLocator())
frequencyx.plot( frequency, 2.0/N*np.abs(frequency_wave[:N//2])**2 )

psdx = figure.add_subplot(313)
psdx.set_title("Power Spectrum")
psdx.set_xlim(left=0, right=150)
#psdx.xaxis.set_major_locator(ticker.NullLocator())
#psdx.yaxis.set_major_locator(ticker.NullLocator())
psdx.psd(temporal_wave, Fs=1000, window=mlab.window_none, pad_to=512)
```

### Spectograma

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

N = 50000
T = 1.0 / 500.0
time = np.linspace(0.0, N*T, N)
frequency = np.linspace(0.0, 1.0/(2.0*T), N//2)

delta_waves = np.sin(0.3*2*np.pi*time) + np.sin(1*2*np.pi*time) + np.sin(4*2*np.pi*time)
theta_waves = np.sin(4.5*2*np.pi*time) + np.sin(6*2*np.pi*time) + np.sin(8*2*np.pi*time)
alpha_waves = np.sin(8.5*2*np.pi*time) + np.sin(10*2*np.pi*time) + np.sin(13*2*np.pi*time)
beta_waves = np.sin(14*2*np.pi*time) + np.sin(20*2*np.pi*time) + np.sin(25*2*np.pi*time) + np.sin(30*2*np.pi*time)
lowgamma_waves = np.sin(31*2*np.pi*time) + np.sin(40*2*np.pi*time) + np.sin(45*2*np.pi*time) + np.sin(50*2*np.pi*time) + np.sin(80*2*np.pi*time)
highgamma_waves = np.sin(80.5*2*np.pi*time) + np.sin(90*2*np.pi*time) + np.sin(100*2*np.pi*time) + np.sin(125*2*np.pi*time) + np.sin(150*2*np.pi*time)

temporal_wave = delta_waves + theta_waves - alpha_waves + beta_waves + lowgamma_waves + highgamma_waves

figure = plt.figure(figsize=(10,10))

plt.specgram(temporal_wave, Fs=300, NFFT=256, noverlap=128, cmap='viridis')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.title('Spectrogram')
plt.show()
```
