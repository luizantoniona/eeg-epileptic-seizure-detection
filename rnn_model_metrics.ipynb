```python
from database.database_metrics import DatabaseMetrics
from ia.metrics.metrics_model import MetricsModel
import ia.metrics.metrics_converter as converter

MODEL = "time_rnn"
DOMAIN = "time"

database = DatabaseMetrics()
db_objects = database.metrics_by_model_and_domain(MODEL, DOMAIN)

time_metrics = []

for db_object in db_objects:
    time_metrics.append(converter.model_from_tuple(db_object))

true_positives = []
true_negatives = []
false_positives = []
false_negatives = []
total_samples = []
accuracy = []
precision = []
sensitivity = []
specificity = []
true_positive_rate = []
false_positive_rate = []
f1_score = []

for time_metric in time_metrics:
    true_positives.append( time_metric.true_positives )
    true_negatives.append( time_metric.true_negatives )
    false_positives.append( time_metric.false_positives )
    false_negatives.append( time_metric.false_negatives )
    total_samples.append( time_metric.total_samples )
    accuracy.append( time_metric.accuracy )
    precision.append( time_metric.precision )
    sensitivity.append( time_metric.sensitivity )
    specificity.append( time_metric.specificity )
    true_positive_rate.append( time_metric.true_positive_rate )
    false_positive_rate.append( time_metric.false_positive_rate )
    f1_score.append( time_metric.f1_score )

final_time_metrics = MetricsModel(MODEL, DOMAIN,
                                  true_positives,
                                  true_negatives,
                                  false_positives,
                                  false_negatives,
                                  total_samples,
                                  accuracy,
                                  precision,
                                  sensitivity,
                                  specificity,
                                  true_positive_rate,
                                  false_positive_rate,
                                  f1_score)

final_time_metrics.show_name()
final_time_metrics.show_mean()
final_time_metrics.show_median()
```
